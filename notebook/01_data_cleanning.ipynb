# **Data Cleaning**

```python
import pandas as pd
```

```python
df = pd.read_csv("../data/data.csv")

print(df.shape)
df.info()
df.describe()
```

```python
# Remove sample where house price < 0.
df = df[df["price"] > 0].copy()

df = df.drop_duplicates()
```

```python
# Save the cleaned data to the data directory.
df.to_csv("../data/cleaned_data.csv", index=False)
```
